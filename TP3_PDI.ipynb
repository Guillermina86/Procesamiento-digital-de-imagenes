## Librerías y funciones a utilizar

```python
# Importamos las librerías
import numpy as np
import matplotlib.pyplot as plt
import imageio as io
```

```python
# Suma
def suma (imagen1, imagen2):
   imagen_suma = imagen1 + imagen2
   return imagen_suma
```

```python
# Resta
def resta (imagen1, imagen2):
    imagen_resta = imagen1 - imagen2
    return imagen_resta
```

```python
# Promedio
def promedio (imagen):
    imagen_prom = imagen / 2
    return imagen_prom
```

```python
# clampeada
def clamp (imagen):
  return np.clip(imagen, 0, 1)
```

```python
# Función para ver una imagen
def get_image(rute):
  imagen = io.imread(rute)
  imagen = imagen.astype(np.int)
  return imagen

# Plotear Imagen
def plot_image(im, title, cmap=None):
  fig, ax = plt.subplots(dpi=100)
  ax.imshow(im, cmap=cmap)
  ax.set_title(title)
  plt.axis('off')
  plt.show()
```

```python
# Obtener caracteristicas de la imagen
def info_image(im):
  print('Dimensiones y número de canales de la imagen:', im.shape)
  print('Tipo de archivo:', im.dtype)
```

# Selección de imágenes

# Suma de imágenes

```python
#@title Seleccione una operación a realizar { run: "auto" }

suma_en_rgb = 'suma_en_rgb'
sumas_promediadas = 'sumas_promediadas'
sumas_clampeadas = 'sumas_clampeadas'
 
Operacion = 'sumas_clampeadas' #@param ["suma_en_rgb", "sumas_promediadas", "sumas_clampeadas"]

if Operacion == 'suma_en_rgb':
  suma_en_rgb = suma(im1, im2)
  plot_image(suma_en_rgb, 'Suma de dos imagenes')
elif Operacion == 'sumas_promediadas':
  sumas_promediadas = promedio(suma(im1/255, im2/255))
  plot_image(sumas_promediadas, 'Imagen suma promediada')
elif Operacion == 'sumas_clampeadas':
  sumas_clampeadas = clamp(suma(im1/255, im2/255))
  plot_image(sumas_clampeadas, 'Imagen suma clampeada')
```

# Resta de imágenes

```python
#@title Seleccione una operación a realizar { run: "auto" }

restas_en_rgb = 'restas_en_rgb'
restas_promediadas = 'restas_promediadas'
restas_clampeadas = 'restas_clampeadas'
 
Operacion = 'restas_clampeadas' #@param ["restas_en_rgb", "restas_promediadas", "restas_clampeadas"]

if Operacion == 'restas_en_rgb':
  restas_en_rgb = resta(im1, im2)
  plot_image(restas_en_rgb, 'Resta de dos imágenes')
elif Operacion == 'restas_promediadas':
  restas_promediadas = promedio(resta(im1/255, im2/255))
  plot_image(restas_promediadas, 'Imagen resta promediada')
elif Operacion == 'restas_clampeadas':
  restas_clampeadas = clamp(resta(im1/255, im2/255))
  plot_image(restas_clampeadas, 'Imagen resta clampeada')
```
